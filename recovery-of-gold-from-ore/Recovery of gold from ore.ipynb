## Описание данных

**Технологический процесс**:

- `Rougher feed` — исходное сырье
- `Rougher additions` (или reagent additions) — флотационные реагенты: Xanthate, Sulphate, Depressant
- `Xanthate` — ксантогенат (промотер, или активатор флотации);
- `Sulphate` — сульфат (на данном производстве сульфид натрия);
- `Depressant` — депрессант (силикат натрия).
- `Rougher process` (англ. «грубый процесс») — флотация
- `Rougher tails` — отвальные хвосты
- `Float banks` — флотационная установка
- `Cleaner process` — очистка
- `Rougher Au` — черновой концентрат золота
- `Final Au` — финальный концентрат золота

**Параметры этапов**:

- `air amount` — объём воздуха
- `fluid levels` — уровень жидкости
- `feed size` — размер гранул сырья
- `feed rate` — скорость подачи

**Наименование признаков**:

Наименование признаков должно быть такое:

[этап].[тип_параметра].[название_параметра]

Возможные значения для блока [этап]:

- `rougher` — флотация
- `primary_cleaner` — первичная очистка
- `secondary_cleaner` — вторичная очистка
- `final` — финальные характеристики

Возможные значения для блока [тип_параметра]:

- `input` — параметры сырья
- `output — параметры продукта
- `state — параметры, характеризующие текущее состояние этапа
- `calculation` — расчётные характеристики

### Проверка правильности расчёта эффективности обогащения

Эффективность обогащения рассчитывается по формуле:

<div>

</div>

где:
- C — доля золота в концентрате после флотации/очистки;
- F — доля золота в сырье/концентрате до флотации/очистки;
- T — доля золота в отвальных хвостах после флотации/очистки.

```python
C = data_train['rougher.output.concentrate_au'] 
F = data_train['rougher.input.feed_au']
T = data_train['rougher.output.tail_au']
```

```python
#Итоговая формула
recovery = (C * (F - T)) / (F * (C - T)) * 100
```

```python
print(f'Средняя абсолютная ошибка MAE составляет {mean_absolute_error(recovery, data_train["rougher.output.recovery"])}')
```

Средняя абсолютная ошибка стремится к 0, следовательно можно сделать вывод о корректности рассчёта эффективности обогащения.

## Модель

### Функция для вычисления sMAPE

```python
def get_smape(target, predict):

    smape = 1 / len(target) * sum(abs(target - predict) / ((abs(target) + abs(predict)) / 2)) * 100
        
    return smape
```

```python
def sMAPE(smape_rougher, smape_final):
    
    sMAPE = 0.25 * smape_rougher + 0.75 * smape_final
    
    return sMAPE
```

```python
smape_scorer = make_scorer(get_smape, greater_is_better=False)
```

Удалим из обучающей выборки столбцы, которые отсутствуют в тестовых данных.

Так как в нашем наборе данных присутствует мультиколлинеарность, будем обучать следующие модели:

### RandomForestRegressor

```python
model = RandomForestRegressor()

#Здесь была сетка с параметрами, но чтобы работало быстрее, я нашла лучшие и оставила только их
param_space = {
    'random_state' : [42],
    'max_depth' : [5],
    'n_estimators' : [50],
    'max_features' : ['log2']
}
```

```python
%%time

rfr_grid = GridSearchCV(model, param_space, scoring=smape_scorer, cv=5, n_jobs=-1)
rfr_grid.fit(features_train, target_train_rougher)

best_rfr = rfr_grid.best_estimator_
best_score_rfr_rougher = rfr_grid.best_score_

print(f'Лучшая метрика sMAPE для процесса rougher: {best_score_rfr_rougher}\nПараметры лучшей модели:\n{best_rfr}')
```

```python
model_rfr_final = RandomForestRegressor(n_estimators=50, max_depth=5, random_state=42, max_features='log2')

best_score_cv_final = cross_val_score(model_rfr_final, features_train, target_train_final, cv=5, scoring=smape_scorer, n_jobs=-1).mean()

print(f'Лучшая метрика sMAPE для процесса final: {best_score_cv_final}')
```

```python
print(f'Итоговая метрика sMAPE для модели {best_rfr} равна {sMAPE(best_score_rfr_rougher, best_score_cv_final)}')
```

### Ridge

```python
%%time

ridge = Ridge(random_state=42)

best_score_ridge_rougher = cross_val_score(ridge, features_train, target_train_rougher, cv=5, scoring=smape_scorer, n_jobs=-1).mean()

print(f'Лучшая метрика sMAPE для процесса final: {best_score_ridge_rougher}')

best_score_ridge_final = cross_val_score(ridge, features_train, target_train_final, cv=5, scoring=smape_scorer, n_jobs=-1).mean()

print(f'Лучшая метрика sMAPE для процесса final: {best_score_ridge_final}')
```

```python
print(f'Итоговая метрика sMAPE для модели {ridge} равна {sMAPE(best_score_ridge_rougher, best_score_ridge_final)}')
```

### Lasso

```python
%%time

lasso = Lasso(random_state=42)

best_score_lasso_rougher = cross_val_score(lasso, features_train, target_train_rougher, cv=5, scoring=smape_scorer, n_jobs=-1).mean()

print(f'Лучшая метрика sMAPE для процесса rougher: {best_score_lasso_rougher}')

best_score_lasso_final = cross_val_score(lasso, features_train, target_train_final, cv=5, scoring=smape_scorer, n_jobs=-1).mean()

print(f'Лучшая метрика sMAPE для процесса final: {best_score_lasso_final}')

```

```python
print(f'Итоговая метрика sMAPE для модели {lasso} равна {sMAPE(best_score_lasso_rougher, best_score_lasso_final)}')
```

### Проверка качества модели на тестовой выборке
