#  Прогнозирование заказов такси

## Описание проекта
Компания собрала исторические данные о заказах такси в аэропортах. Чтобы привлекать больше водителей в период пиковой нагрузки, нужно спрогнозировать количество заказов такси на следующий час. Постройте модель для такого предсказания.

## Цель проекта
Построить модель, которая прогнозирует количество заказов такси на следующий час. Значение метрики *RMSE* на тестовой выборке должно быть не больше 48.

## Описание данных
- `num_orders` - количество заказов

## План работы
1. Загрузить данные и выполнить их ресемплирование по одному часу.
2. Проанализировать данные.
3. Обучить разные модели с различными гиперпараметрами. Сделать тестовую выборку размером 10% от исходных данных.
4. Проверить данные на тестовой выборке и сделать выводы.

```python
import pandas as pd
import matplotlib.pyplot as plt 
import catboost as cb

from numpy import logspace
from statsmodels.tsa.seasonal import seasonal_decompose
from statsmodels.tsa.stattools import adfuller
from statsmodels.graphics.tsaplots import plot_acf
from sklearn.model_selection import cross_val_score, GridSearchCV , train_test_split, TimeSeriesSplit 
from sklearn.linear_model import LinearRegression 
from sklearn.ensemble import RandomForestRegressor 
from sklearn.metrics import make_scorer, mean_squared_error 
```

## Подготовка

```python
try:
    data = pd.read_csv('/datasets/taxi.csv', index_col=[0], parse_dates=[0])
except:
    data = pd.read_csv(r'C:\Users\Acer\Documents\Практикум\taxi.csv', index_col=[0], parse_dates=[0])
    
data.head()
```

Отсортируем даты в порядке возрастания и проверим их хронологический порядок.

```python
data = data.sort_index()
data.index.is_monotonic
```

В датасете представлены данные за 184 дня работы.

В среднем за день поступает 2027 заказов.

### Cтационарность

Дляоценки стационарности применим Dickey-Fuller test.

*Гипотезы*:

Нулевая гипотеза предполагает, что процесс нестационарный.


Альтернативная гипотеза соответственно говорит об обратном.

Порог возьмём 0.05

```python
adf_test = adfuller(data['num_orders'], regression='ct')

print('p-value = ' + str(adf_test[1]))
```

В нашем случае нулевая гипотеза отвергается, т.к. значение сильно меньше порога, следовательно график стационарен.

### Автокорреляция

```python
plot_acf(data)
plt.show()
```

**Выводы**:
Данные разбиты по часам
1. Можно говорить о ежесуточной сезонности, 23, 24, 25 лаг имеет более сильную корреляцию с 1 лагом.
2. Так же можно заметить достаточно сильную автокорреляцию, т.к. синяя зона небольшая и все переменные выходят за её пределы.
3. Все лаги находятся выше 0, что может говорить нам о наличии тренда. Хотя в графиках этого тренда не наблюдалось.

Добавим в данные новые признаки:
- `day` - день
- `month` - месяц
- `dayofweek` - день недели

```python
def make_features(data, max_lag, rolling_mean_size):
    data['dayofweek'] = data.index.dayofweek
    data['hour'] = data.index.hour
    
    for lag in range(1, max_lag + 1):
        data['lag_{}'.format(lag)] = data['num_orders'].shift(lag)

    data['rolling_mean'] = data['num_orders'].shift().rolling(rolling_mean_size).mean()
```

```python
make_features(data, 24, 48)
data = data.dropna()
data.head()
```

```python
train, test = train_test_split(data, shuffle=False, test_size=0.1, random_state=42)

X_train = train.drop(['num_orders'], axis=1)
y_train = train['num_orders']
X_test = test.drop(['num_orders'], axis=1)
y_test = test['num_orders']
```

```python
X_test.info()
```

## Обучение

С помощью функции make_scorer создадим метрику RMSE

```python
def RMSE(target, predict):
    return (mean_squared_error(target, predict))**0.5
   
RMSE_scorer = make_scorer(RMSE, greater_is_better=False)
```

Будем обучать несколько моделей:
1. Linear Regression
2. Random Forest
3. Catboost

### Linear Regression

```python
%%time
tscv = TimeSeriesSplit(n_splits=5)

model_reg = LinearRegression()

model_reg.fit(X_train, y_train)

model_reg_score = pd.Series(cross_val_score(model_reg, X_train, y_train, scoring=RMSE_scorer, cv=tscv)).mean()*(-1)
model_reg_score
```

### Random Forest

```python
%%time

model_rf = RandomForestRegressor(random_state=42)

param = { 'n_estimators': range(10, 50, 10),
          'max_depth': range(1, 15, 5),
          'min_samples_leaf': range(1, 5),
          'min_samples_split': range(2, 8, 2) }


grid_rf = GridSearchCV(model_rf, param, scoring=RMSE_scorer, cv=tscv)
grid_rf.fit(X_train, y_train)
```

```python
grid_rf.best_score_ * (-1)
```

### Catboost

```python
%%time
model_cbr = cb.CatBoostRegressor(random_state=42)

param = {'learning_rate': logspace(-2, 0, 5),
                'iterations': range(20, 80, 20),
                'depth': range(5,10,2)}

grid_cbr = GridSearchCV(model_cbr, param, scoring=RMSE_scorer, cv=tscv)
grid_cbr.fit(X_train, y_train, verbose=False)
```

```python
grid_cbr.best_score_ * (-1)
```

**Вывод**:
Лучшее качество показала модель Catboost, её и будем использовать для финального тестирования.

## Тестирование

```python
predictions = grid_cbr.predict(X_test)
rmse = RMSE(y_test, predictions)
rmse
```

```python
predictions = pd.DataFrame(predictions, index=X_test.index)
```

```python
plt.figure(figsize=(18, 6))
plt.plot(y_test, color = "b")
plt.plot(predictions, color = "green")
 
# заголовок и подписи к осям
plt.title("Тестовая выборка и тестовый прогноз")
plt.ylabel('Количество заказов')
plt.xlabel('Даты, часы')
 
# добавим сетку и легенду
plt.grid()
plt.legend(('Тестовая выборка', 'Тестовый прогноз'))
 
plt.show()
```
